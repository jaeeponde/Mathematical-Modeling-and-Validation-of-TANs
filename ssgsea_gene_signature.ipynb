```python
import pandas as pd
df1 = pd.DataFrame(ss1.res2d)
df2 = pd.DataFrame(ss2.res2d)
```

```python
df1 = df1.sort_values(by='Name')
df1
```

```python
df2 = df2.sort_values(by='Name')
df2
```

```python
df1.columns
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your dataframes df1 and df2

# Extract the 'NES' columns from both dataframes

nes_df1 = df1[df1['Term'] == "HALLMARK_EPITHELIAL_MESENCHYMAL_TRANSITION"]['NES']
nes_df2 = df2[df2['Term'] == "MARTINELLI_IMMATURE_NEUTROPHIL_DN"]['NES']

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(nes_df1, nes_df2, alpha=0.5)
plt.title('Scatter Plot of NES between df1 and df2')
plt.xlabel('EMT')
plt.ylabel('IMMATURE NEUTROPHILS')
plt.grid(True)
plt.show()

```

```python
from scipy.stats import pearsonr
```

```python
correlation_coefficient, p_value = pearsonr(df1['NES'], df2['NES'])

print(f"Pearson correlation coefficient between '{'NES Neutrophils'}' and '{'NES EMT'}': {correlation_coefficient}")
print(f"P-value: {p_value}")
```
